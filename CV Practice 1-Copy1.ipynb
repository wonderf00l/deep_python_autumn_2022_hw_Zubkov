## Занятие 1. Получение и вывод изображения с помощью OpenCV

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
from PIL import Image

%matplotlib inline
```

**Чтение изображения из файла**

```python
opencv_image = cv2.imread('lena.png')
```

**Вывод изображения**

```python
def show_image(image):
    pil_img = Image.fromarray(image)
    return pil_img
```

```python
show_image(opencv_image)
```

**Представление изображения**

**Перевод цветовых пространств**

Перевод из BGR в RGB

```python
bgr_image = opencv_image
```

```python
rgb_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2RGB)
```

```python
show_image(rgb_image)
```

```python
print(bgr_image[0, 0])
print(rgb_image[0, 0])
```

Перевод в черно-белый формат (одноканальный)

```python
grayscale_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2GRAY)
```

```python
show_image(grayscale_image)
```

Перевод в HSV (Hue Saturation Value)

```python
hsv_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2HLS)
```

```python
show_image(hsv_image)
```

```python
print(hsv_image[0, 0])
```

### Кодирование изображений

Кодирование цветного JPEG

```python
_, jpeg_encoded_data = cv2.imencode(".jpg", opencv_image)
```

```python
print(jpeg_encoded_data)
```

```python
decoded_image = cv2.imdecode(jpeg_encoded_data, cv2.IMREAD_COLOR)
```

```python
show_image(decoded_image)
```

Кодирование черно-белого PNG

```python
_, gray_png_encoded_data = cv2.imencode(".png", grayscale_image)
```

```python
print(gray_png_encoded_data)
```

```python
gray_decoded_image = cv2.imdecode(gray_png_encoded_data, cv2.IMREAD_GRAYSCALE)
```

```python
show_image(gray_decoded_image)
```

Сравнение размеров изображений

```python
_, png_data = cv2.imencode(".png", opencv_image)
_, jpeg_data = cv2.imencode(".jpg", opencv_image)

print("{} байт - без сжатия".format(opencv_image.size))
print("{} байт - со сжатием без потерь (PNG)".format(png_data.size))
print("{} байт - со сжатием с потерями (JPEG)".format(jpeg_data.size))
```
